```python
window_size = 40
data_arr = normalized_data.to_numpy()
X = []
y = []
for i in range(0,len(data_arr)-window_size):
    X.append(data_arr[i:i+window_size])
    y.append(data_arr[i+window_size,0])
X = np.array(X)
y = np.array(y)
print("X shape: ",X.shape)
print("y shape: ",y.shape)
```

```python
index = int(len(X) * .9)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import LSTM,Dense,Dropout,GRU,Flatten
```

```python
model = Sequential()
```

```python
model.add(LSTM(64,input_shape=X_train[0].shape,return_sequences=True))
model.add(Dropout(.2))
```

```python
model.add(LSTM(64,return_sequences=True))
model.add(Dropout(.2))
```

```python
model.add(LSTM(64,return_sequences=True))
model.add(Dropout(.2))
```

```python
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(.2))
```

```python
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
import tensorflow.keras.backend as K
import tensorflow as tf
def smape_loss(y_true, y_pred):
    epsilon = 0.1
    summ = K.maximum(K.abs(y_true) + K.abs(y_pred) + epsilon, 0.5 + epsilon)
    smape = K.abs(y_pred - y_true) / summ * 2.0
    return smape * 100.0
model.compile(optimizer=tf.keras.optimizers.RMSprop(1e-7),loss='mse',metrics=['mae','mse','mean_absolute_percentage_error',smape_loss])
```

```python
history = model.fit(X_train,y_train,epochs=100
                    ,batch_size=128,shuffle=True,use_multiprocessing=True,validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test,y_test)
```
